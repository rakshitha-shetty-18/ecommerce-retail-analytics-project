## Day 4 - Data Cleaning
**Date:** 11-June-2026


```python
# Import libraries
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_excel(r'C:\Users\User\Desktop\DataAnalysesProject\data\raw\Online Retail.xlsx')

print("Dataset loaded!")
print("Shape:", df.shape)
```

```python
# Check how many cancelled orders exist BEFORE cleaning
print("Before removing cancellations:", df.shape)
print("Cancelled orders (Quantity < 0):", (df['Quantity'] < 0).sum())

# Remove rows where Quantity is less than 0
df = df[df['Quantity'] > 0]

print("After removing cancellations:", df.shape)
```

```python
# Check how many zero price rows exist
print("Zero UnitPrice rows:", (df['UnitPrice'] <= 0).sum())

# Remove rows where UnitPrice is 0 or negative
df = df[df['UnitPrice'] > 0]

print("After removing zero prices:", df.shape)
```

```python
# Check missing CustomerIDs
print("Missing CustomerIDs:", df['CustomerID'].isnull().sum())

# Remove rows where CustomerID is blank
df = df.dropna(subset=['CustomerID'])

print("After removing missing CustomerIDs:", df.shape)
```

```python
# List of junk descriptions to remove
junk = ['AMAZON FEE', 'Manual', 'SAMPLES', 'Damaged', 
        'Bank Charges', 'POSTAGE', 'DOTCOM POSTAGE',
        'Adjust bad debt', 'CRUK Commission']

# Check how many junk rows exist
print("Junk rows:", df['Description'].isin(junk).sum())

# Remove junk rows
df = df[~df['Description'].isin(junk)]

print("After removing junk:", df.shape)
```

```python
# Fix InvoiceDate from text to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Fix CustomerID from float to integer
df['CustomerID'] = df['CustomerID'].astype(int)

# Verify the fix
print("Data types after fixing:")
print(df[['InvoiceDate', 'CustomerID']].dtypes)
```

```python
# Create TotalRevenue column
df['TotalRevenue'] = df['Quantity'] * df['UnitPrice']

# Extract Month from InvoiceDate
# dt.month gives month number (1=Jan, 2=Feb etc)
df['Month'] = df['InvoiceDate'].dt.month

# Extract Year from InvoiceDate
df['Year'] = df['InvoiceDate'].dt.year

# Extract DayOfWeek from InvoiceDate
# 0=Monday, 1=Tuesday ... 6=Sunday
df['DayOfWeek'] = df['InvoiceDate'].dt.dayofweek

# Verify new columns
print("New columns added!")
print(df[['Quantity', 'UnitPrice', 'TotalRevenue', 
          'Month', 'Year', 'DayOfWeek']].head())
```

```python
# 1. Check outliers in Quantity
print("Quantity stats:")
print("Max:", df['Quantity'].max())
print("Min:", df['Quantity'].min())
print("Mean:", round(df['Quantity'].mean(), 2))

# 2. np.percentile - finds the value at a given percentage
# 99th percentile means - 99% of values are below this number
q99 = np.percentile(df['Quantity'], 99)
print("\n99th percentile of Quantity:", q99)

# 3. np.clip - caps values between a minimum and maximum
# Any Quantity above 99th percentile gets capped
df['Quantity_clipped'] = np.clip(df['Quantity'], 1, q99)
print("Quantity clipped successfully!")

# 4. np.log - log transform of TotalRevenue
# Useful for visualization when data is skewed
df['LogRevenue'] = np.log(df['TotalRevenue'] + 1)
print("Log Revenue created!")

# Verify
print("\nSample:")
print(df[['Quantity', 'Quantity_clipped', 
          'TotalRevenue', 'LogRevenue']].head())
```

```python
# Define output path
output_path = r'C:\Users\User\Desktop\DataAnalysesProject\data\clean\online_retail_clean.csv'

# Save cleaned dataframe to CSV
df.to_csv(output_path, index=False)

print("✅ Clean data saved successfully!")
print("Location:", output_path)
print("Final shape:", df.shape)
print("Columns:", df.columns.tolist())
```

## Day 5 - RFM Analysis
**Date:** 11-June-2026

```python
import pandas as pd
import numpy as np

# Load CLEAN data today - not raw!
df = pd.read_csv(r'C:\Users\User\Desktop\DataAnalysesProject\data\clean\online_retail_clean.csv')

# Fix InvoiceDate back to datetime after loading CSV
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

print("Clean data loaded!")
print("Shape:", df.shape)
print("Date range:", df['InvoiceDate'].min(), "to", df['InvoiceDate'].max())
```

```python
# Reference date = last date in dataset + 1 day
reference_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
print("Reference date:", reference_date)

# Calculate RFM for each customer
rfm = df.groupby('CustomerID').agg(
    Recency   = ('InvoiceDate', lambda x: (reference_date - x.max()).days),
    Frequency = ('InvoiceNo', 'nunique'),
    Monetary  = ('TotalRevenue', 'sum')
).reset_index()

print("\nRFM table created!")
print("Shape:", rfm.shape)
print("\nFirst 5 customers:")
print(rfm.head())
```

```python
# Calculate percentile boundaries using NumPy
r_percentiles = np.percentile(rfm['Recency'],   [25, 50, 75])
f_percentiles = np.percentile(rfm['Frequency'], [25, 50, 75])
m_percentiles = np.percentile(rfm['Monetary'],  [25, 50, 75])

print("Recency boundaries:",   r_percentiles)
print("Frequency boundaries:", f_percentiles)
print("Monetary boundaries:",  m_percentiles)

# Score Recency - REVERSED (low recency = better score)
rfm['R_Score'] = np.digitize(rfm['Recency'],
                 bins=r_percentiles, right=True)
rfm['R_Score'] = 4 - rfm['R_Score']  # Reverse the score

# Score Frequency - normal (high frequency = better score)
rfm['F_Score'] = np.digitize(rfm['Frequency'],
                 bins=f_percentiles, right=True)

# Score Monetary - normal (high monetary = better score)  
rfm['M_Score'] = np.digitize(rfm['Monetary'],
                 bins=m_percentiles, right=True)

print("\nRFM Scores added!")
print(rfm.head())
```

```python
# Create combined RFM score
rfm['RFM_Score'] = rfm['R_Score'] + rfm['F_Score'] + rfm['M_Score']

# Create segments based on scores
def assign_segment(row):
    if row['R_Score'] >= 3 and row['F_Score'] >= 2:
        return 'Recent Active'
    elif row['F_Score'] >= 3 and row['M_Score'] >= 2:
        return 'Frequent Buyers'
    elif row['M_Score'] >= 3:
        return 'High Value'
    elif row['R_Score'] <= 1 and row['F_Score'] <= 1:
        return 'Lost Customers'
    else:
        return 'Average Customers'

# Apply segment function to each row
rfm['Segment'] = rfm.apply(assign_segment, axis=1)

# Count customers in each segment
print("Customer Segments:")
print(rfm['Segment'].value_counts())
print("\nTotal customers:", len(rfm))
```

```python
# Save RFM table to clean folder
rfm_path = r'C:\Users\User\Desktop\DataAnalysesProject\data\clean\rfm_segments.csv'

rfm.to_csv(rfm_path, index=False)

print("✅ RFM data saved!")
print("Columns:", rfm.columns.tolist())
print("\nFinal RFM table sample:")
print(rfm.head(10))
```

```python
# Statistical summary of RFM
print("=" * 50)
print("RFM STATISTICAL SUMMARY")
print("=" * 50)

print("\n📊 RECENCY (days since last purchase):")
print(f"  Best customer:  {rfm['Recency'].min()} days ago")
print(f"  Worst customer: {rfm['Recency'].max()} days ago")
print(f"  Average:        {round(rfm['Recency'].mean(), 1)} days")
print(f"  Median:         {rfm['Recency'].median()} days")

print("\n📊 FREQUENCY (number of orders):")
print(f"  Most orders:    {rfm['Frequency'].max()}")
print(f"  Least orders:   {rfm['Frequency'].min()}")
print(f"  Average orders: {round(rfm['Frequency'].mean(), 1)}")
print(f"  Median orders:  {rfm['Frequency'].median()}")

print("\n📊 MONETARY (total spent):")
print(f"  Highest spender: £{round(rfm['Monetary'].max(), 2)}")
print(f"  Lowest spender:  £{round(rfm['Monetary'].min(), 2)}")
print(f"  Average spend:   £{round(rfm['Monetary'].mean(), 2)}")
print(f"  Median spend:    £{round(rfm['Monetary'].median(), 2)}")

print("\n📊 SEGMENT SUMMARY:")
segment_summary = rfm.groupby('Segment').agg(
    Count    = ('CustomerID', 'count'),
    Avg_Recency   = ('Recency', 'mean'),
    Avg_Frequency = ('Frequency', 'mean'),
    Avg_Monetary  = ('Monetary', 'mean')
).round(1)
print(segment_summary)
```

```python
import pandas as pd
import numpy as np

# Load clean data
df = pd.read_csv(r'C:\Users\User\Desktop\DataAnalysesProject\data\clean\online_retail_clean.csv')

# Fix datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

print("Clean data loaded!")
print("Shape:", df.shape)
```

```python
# Add Month Name column
# %B gives full month name (January, February etc)
# %b gives short month name (Jan, Feb etc)
df['MonthName'] = df['InvoiceDate'].dt.strftime('%b')

# Add Month-Year column (useful for time series charts)
df['MonthYear'] = df['InvoiceDate'].dt.strftime('%b-%Y')

print("Month columns added!")
print(df[['InvoiceDate', 'Month', 'MonthName', 'MonthYear']].head(10))
```
